```python
import networkx as nx
import matplotlib.pyplot as plt

def plot_topology(adj_matrix):
    G = nx.Graph()
    for i in range(len(adj_matrix)):
        for j in range(i, len(adj_matrix[i])):
            if adj_matrix[i][j] == 1:
                G.add_edge(i, j)

    plt.figure(figsize=(8, 6))
    nx.draw(G, with_labels=True, node_color='lightblue', edge_color='gray', node_size=2000, font_size=12)
    plt.title("Network Topology Visualization")
    plt.show()

plot_topology(sample_adj_matrix)
```
